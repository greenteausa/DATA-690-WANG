## 3. Download and read the documentation - https://collegescorecard.ed.gov/assets/FullDataDocumentation.pdf

```python
# Import libraries
import requests
import PyPDF2 
import pandas as pd
import urllib.request
import zipfile
```

```python
# I use this link as my refefrence. https://www.geeksforgeeks.org/downloading-files-web-using-python/
url = 'https://collegescorecard.ed.gov/assets/FullDataDocumentation.pdf'
r = requests.get(url, stream=True)

with open('/tmp/FullDataDocumentation.pdf', 'wb') as fd:
    for chunk in r.iter_content(chunk_size=1024):
        fd.write(chunk)
```

```python
# I use the this link as my reference. https://www.geeksforgeeks.org/working-with-pdf-files-in-python/
pdfFileObj = open('/tmp/FullDataDocumentation.pdf', 'rb')
pdfReader = PyPDF2.PdfFileReader(pdfFileObj)
pageObj = pdfReader.getPage(0)
print(pageObj.extractText()) 
```

```python
pdfFileObj.close() 
```

## 4. Download the dataset - https://ed-public-download.app.cloud.gov/downloads/CollegeScorecard_Raw_Data.zip (265MB Zip File)

```python
# I use this as my reference. https://stackoverflow.com/questions/6861323/download-and-unzip-file-with-python/30327074
url = 'https://ed-public-download.app.cloud.gov/downloads/CollegeScorecard_Raw_Data.zip '
filehandle, _ = urllib.request.urlretrieve(url)
zip_file_object = zipfile.ZipFile(filehandle, 'r')
first_file = zip_file_object.namelist()[0]
file = zip_file_object.open(first_file)
content = file.read()
```
