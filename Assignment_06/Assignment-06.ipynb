### Write a function named cleanse_year() that takes a string as input and removes any asterisks in the string and return the cleansed string. Test the function by using test strings (for example, "1989*", "20*10")

```python
def cleanse_year(item):
    
    item = item.replace('*','')  
        
    return item
```

```python
cleanse_year( "1989*")
```

```python
cleanse_year( "20*10")
```

### Cleanse the column "Census Year" using the function cleanse_year() and Pandas's apply() function.

```python
df1["Census Year"] = df1["Census Year"].apply(cleanse_year)
df1
```

### Repeat this process for column "Total Population", "Census Cost", "Average Cost Per Person". (define, test, and apply the function)

```python
def cost_list(item):
    
    item = item.strip('$')  # drop the '$'.
    item = item.replace(',','')# drop the ','.
    if 'Billion' in item: # change the 'Billion' to 1000000000 and change it to float then can multiple with 10**9,  and int then str to join easily. 
        item = item.strip('Billion')
        item = str(int(float(item) *1000000000))
        
    return item
```

```python
df1["Census Cost"] = df1["Census Cost"].apply(cost_list)
df1.head()
```

```python
df1
```

```python
def Average_Cost(item):
    
    item = item.strip('cents')
   
    if '$' in item:
        item = item.strip('$') # drop '$'.
        item = str(float(item) *100) # float to multiple and then string then to join easily.
    return item
```

```python
df1["Average Cost Per Person"] = df1["Average Cost Per Person"].apply(Average_Cost)
df1
```

### After all columns are cleansed, save the clenased dataframe to a file named "census_cost_cleansed.csv" using CSV format.
